```python
import pandas as pd
from geopy.geocoders import Nominatim
import tqdm
geolocator = Nominatim(user_agent="blah")

def fun(row):
    return row['Zone'] if '/' not in row['Zone'] else row['Zone'].split('/')[0]
    
df = pd.read_csv("taxi+_zone_lookup.csv")
df = df[df['Borough'] != 'Unknown']
df['loc'] = df.apply(fun, axis=1)

df['lat'] = 0
df['lng'] = 0

def get_lat_long(row):
    if row['lat'] !=0:
        return row
    location = None
    for suffix in [', New York', '', f", {row['Borough']}", ', New Jersey']:
        try:
            location = geolocator.geocode(row['loc'] + suffix)
            if location is not None and not (38 < location.latitude < 47 and -80 < location.longitude < -71):
                location = None
        except:
            pass
        if location is not None:
            break
        
    if location is not None:
        row['lat'] = location.latitude
        row['lng'] = location.longitude
    else:
        row['lat'] = 0.0
        row['lng'] = 0.0
    return row

for i, row in tqdm.tqdm(df.iterrows()):
    df.iloc[i] = get_lat_long(row)
df
```

```python
df.to_csv('taxi_zone_updated.csv', index=False)
```
